```python
# Dependencies and Setup
import pandas as pd
```

```python
# Read in csv file and display data
citibike11 = pd.read_csv("202311-citibike-tripdata.csv")
citibike11.head()
```

```python
# Read in csv file and display data
citibike12 = pd.read_csv("202312-citibike-tripdata.csv")
citibike12.head()
```

```python
#merge Datasets
citibike = citibike11.merge(citibike12, how="outer")
citibike
```

```python
citibike_clean = citibike.dropna(how="any")
citibike_clean
```

```python
citibike_clean.info()
```

```python
citibike_clean.to_csv('citibike2023.csv', index=False)
```
